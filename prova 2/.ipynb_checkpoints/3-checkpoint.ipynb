Importando módulos necessários

```python
import pandas as pd
import numpy as np
from sklearn import neighbors, metrics, preprocessing
np.set_printoptions(suppress=True)
```

Abrindo o arquivo penguins_size.csv como dataframe chamado originalDF

```python
originalDF = pd.read_csv('Mall_Customers.csv', encoding = "ISO-8859-1")
```

Codificando os campos que não eram compostos por números

```python
label_encoder = preprocessing.LabelEncoder()

encodedDF = originalDF.copy()
encodedDF['Genre']= label_encoder.fit_transform(encodedDF['Genre'])

encodedDF = encodedDF[~np.isnan(encodedDF).any(axis=1)]
encodedDF.reset_index(drop=True, inplace=True)
encodedDF
```

Criando um dataframe padronizado

```python
from sklearn.preprocessing import StandardScaler

scaledArr = StandardScaler().fit_transform(encodedDF.loc[:, ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].values)
scaledDF = pd.DataFrame(data = scaledArr, columns = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)'])
scaledDF
```

Incluindo no dataframe padronizado a espécie

```python
scaledDF = pd.concat([scaledDF, encodedDF[['CustomerID', 'Genre']]], axis = 1)
scaledDF
```

Separando o rótulo dos componentes, sendo y o rótulo

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

principalComponents2d = pca.fit_transform(scaledDF)
principalDf2d = pd.DataFrame(data = principalComponents2d, columns = ['Componente Principal 1', 'Componente Principal 2'])
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(principalDf2d)
```

Aplicando oversampling na base para balancear o número de classes

```python
import matplotlib.pyplot as plt

plt.plot(encodedDF[['Annual Income (k$)']], encodedDF[['Spending Score (1-100)']])
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)

principalComponents3d = pca.fit_transform(scaledDF)
principalDf3d = pd.DataFrame(data = principalComponents3d, columns = ['Componente Principal 1', 'Componente Principal 2', 'Componente Principal 3'])

finalDf3d = pd.concat([principalDf3d, encodedDF[['Genre']]], axis = 1)
finalDf3d
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,10))

axis = fig.add_subplot(111, projection='3d')
 
axis.set_facecolor('xkcd:black')

axis.scatter(finalDf3d['Componente Principal 1'],finalDf3d['Componente Principal 2'],finalDf3d['Componente Principal 3'], c=finalDf3d['Genre'], cmap='Set3')
axis.set_xlabel("PC1", fontsize=10)
axis.set_ylabel("PC2", fontsize=10)
axis.set_zlabel("PC3", fontsize=10)
```
