Importando os módulos principais

```python
import pandas as pd
import numpy as np
from sklearn import neighbors, metrics, preprocessing
import seaborn as sb
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
np.set_printoptions(suppress=True)
```

Abrindo o arquivo Mall_Customers.csv como dataframe chamado originalDF

```python
originalDF = pd.read_csv('Mall_Customers.csv', encoding = "ISO-8859-1")
```

Codificando o campo e que não era composto por números e limpando a base de campos nulos, resultando em um dataframe encodedDF

```python
label_encoder = preprocessing.LabelEncoder()

encodedDF = originalDF.copy()
encodedDF['Genre']= label_encoder.fit_transform(encodedDF['Genre'])

encodedDF = encodedDF[~np.isnan(encodedDF).any(axis=1)]
encodedDF.reset_index(drop=True, inplace=True)
encodedDF
```

Plotando grafico do dataframe encodedDF

```python
sb.pairplot(encodedDF)
```

Tirando variável CustomerID do dataframe

```python
X = encodedDF.drop('CustomerID', axis=1)
X.head()
```

Plotando gráfico para definir numero de clusters indicados pelo método do cotovelo

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', n_init = 10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title("Método do Cotovelo")
plt.xlabel("Número de Clusters")
plt.ylabel("Within-Cluster-Sum-of-Squares")
plt.show()
```

Definindo X como todo o dataframe

```python
X = np.array(encodedDF)
```

Executando o KMeans usando como numero de clusters 2

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)
kmeans.labels_

encodedDF['K-classes'] = kmeans.labels_
encodedDF

sb.pairplot(encodedDF,hue='K-classes')
```

Executando o KMeans usando como numero de clusters 4

```python
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(X)
kmeans.labels_

encodedDF['K-classes'] = kmeans.labels_
encodedDF

sb.pairplot(encodedDF,hue='K-classes')
```

Executando o KMeans usando como numero de clusters 6

```python
kmeans = KMeans(n_clusters=6, random_state=0)
kmeans.fit(X)
kmeans.labels_

encodedDF['K-classes'] = kmeans.labels_
encodedDF

sb.pairplot(encodedDF,hue='K-classes')
```

Montando dendrograma para usar no metodo hierarquico

```python
from scipy.cluster.hierarchy import dendrogram, linkage

dendrogram(linkage(X))
plt.show()
```

Montando novo dendrograma usando ward de linkage e definindo nivel maximo como 4

```python
dendrogram(linkage(X, 'ward'), truncate_mode='level', p=4)
plt.show()
```

Utilizando metodo hierarquico com linkage ward e numero de clusters = 2 e plotando gráficos.

```python
from sklearn.cluster import AgglomerativeClustering

hc_model = AgglomerativeClustering(affinity='euclidean', linkage='ward')
clusters = hc_model.fit_predict(X)

encodedDF = encodedDF.drop('K-classes', 1)
encodedDF['Clusters'] = clusters

sb.pairplot(encodedDF,hue='Clusters')
```

Utilizando metodo hierarquico com linkage ward e numero de clusters = 3 e plotando gráficos.

```python
hc_model = AgglomerativeClustering(n_clusters= 3, affinity='euclidean', linkage='ward')
clusters = hc_model.fit_predict(X)

encodedDF['Clusters'] = clusters
sb.pairplot(encodedDF,hue='Clusters')
```

Utilizando metodo hierarquico com linkage ward e numero de clusters = 5 e plotando gráficos.

```python
hc_model = AgglomerativeClustering(n_clusters= 5, affinity='euclidean', linkage='ward')
clusters = hc_model.fit_predict(X)

encodedDF['Clusters'] = clusters
sb.pairplot(encodedDF,hue='Clusters')
```

Utilizando metodo hierarquico com linkage complete e numero de clusters = 2 e plotando gráficos.

```python
hc_model = AgglomerativeClustering(affinity='euclidean', linkage='complete')
clusters = hc_model.fit_predict(X)

encodedDF['Clusters'] = clusters
sb.pairplot(encodedDF,hue='Clusters')
```

Utilizando metodo hierarquico com linkage complete e numero de clusters = 3 e plotando gráficos.

```python
hc_model = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='complete')
clusters = hc_model.fit_predict(X)

encodedDF['Clusters'] = clusters
sb.pairplot(encodedDF,hue='Clusters')
```

Utilizando metodo hierarquico com linkage complete e numero de clusters = 5 e plotando gráficos.

```python
hc_model = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete')
clusters = hc_model.fit_predict(X)

encodedDF['Clusters'] = clusters
sb.pairplot(encodedDF,hue='Clusters')
```
